```python
import numpy as np
df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]=df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']].replace(0,np.NaN)
```

```python
import missingno as msno
msno.bar(df,color='purple')
```

```python
#filling up the missing value
#with the median values
def median_target(var):
  temp=df[df[var].notnull()]
  temp=temp[[var,'Outcome']].groupby(['Outcome'])[[var]].median().reset_index()
  return temp
```

```python
columns=df.columns
columns=columns.drop('Outcome')
for i in columns:
  median_target(i)
  df.loc[(df['Outcome']==0)&(df[i].isnull()),i]=median_target(i)[i][0]
  df.loc[(df['Outcome']==1)&(df[i].isnull()),i]=median_target(i)[i][1]
```

```python
#outliers finding
#IQR (Interquartile range)
#Q1-1.5*IQR    IQR      Q3+1.5*IQR
#        Q1  .........   Q3          bell curve
#   24.65%  .         . 24.65%
#         ..     50%    ..
#     ....               ....
#-------------------------------------
# IQR=Q3-Q1
# lower limit = Q1-1.5*IQR
# upper limit = Q3-1.5*IQR
#Q3 = 24.65--->25+50=75
#Q2 = 24.65--->25
for feature in df:
  Q1 = df[feature].quantile(0.25)
  Q3 = df[feature].quantile(0.75)
  IQR=Q3-Q1
  lower=Q1-1.5*IQR
  upper=Q3+1.5*IQR
  if df[(df[feature]>upper)].any(axis=None):
    print(feature,' Yes')
  else:
    print(feature, ' No')
```

##**Model Training before Feature Engineering**

```python
y_y=df['Outcome']
x_x=df.drop(['Outcome'],axis=1)
```

```python
#train,test and splitting of the data
from sklearn.model_selection import train_test_split
x_train_f,x_test_f,y_train_f,y_test_f=train_test_split(x_x,y_y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import scale, StandardScaler
scaler=StandardScaler()
x_train_f=scaler.fit_transform(x_train_f)
x_test_f=scaler.transform(x_test_f)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from scipy.interpolate import make_interp_spline

train_sizes = np.linspace(0.1, 0.9, 10)

models = {
    'KNN': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42)
}

train_accuracies = {name: [] for name in models}
test_accuracies = {name: [] for name in models}

for train_size in train_sizes:
    x_train_subset, _, y_train_subset, _ = train_test_split(x_train_f, y_train_f, train_size=train_size, random_state=42)

    for name, model in models.items():
        model.fit(x_train_subset, y_train_subset)

        train_pred = model.predict(x_train_subset)
        test_pred = model.predict(x_test_f)

        train_acc = accuracy_score(y_train_subset, train_pred)
        test_acc = accuracy_score(y_test_f, test_pred)

        train_accuracies[name].append(train_acc)
        test_accuracies[name].append(test_acc)

for name in models:
    final_train_acc = train_accuracies[name][-1]
    final_test_acc = test_accuracies[name][-1]
    print(f'Model: {name}, Final Training Accuracy: {final_train_acc:.4f}, Final Testing Accuracy: {final_test_acc:.4f}')

plt.figure(figsize=(12, 8))

colors = ['blue', 'green', 'red', 'purple']
markers_train = ['o', 's', '^', 'D']
markers_test = ['x', 'P', '*', 'v']

for i, (name, color) in enumerate(zip(models, colors)):
    x_smooth = np.linspace(train_sizes.min() * len(x_train_f), train_sizes.max() * len(x_train_f), 300)
    train_smooth = make_interp_spline(train_sizes * len(x_train_f), train_accuracies[name])(x_smooth)
    test_smooth = make_interp_spline(train_sizes * len(x_train_f), test_accuracies[name])(x_smooth)

    plt.plot(x_smooth, train_smooth, label=f'{name} - Training', color=color, marker=markers_train[i], markevery=30)
    plt.plot(x_smooth, test_smooth, label=f'{name} - Validation', linestyle='--', color=color, marker=markers_test[i], markevery=30)

plt.title('Effect of Training Set Size on Overfitting (KNN, Decision Tree, Random Forest, Logistic Regression)', fontsize=16)
plt.xlabel('Number of Training Samples', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=12)
plt.tight_layout()
plt.show()
```

##**Feature Engineering**

```python
def set_insulin(row):
  if row['Insulin']>=16 and row['Insulin']<=166:
    return 'Normal'
  else:
    return 'Abnormal'
```

```python
df = df.assign(NewInsulinScore=df.apply(set_insulin,axis=1))
df.head()
```

```python
NewGlucose=pd.Series(['Low', 'Normal', 'Prediabetes', 'Diabetes'],dtype='category')
df['NewGlucose']='Unknown'
df.loc[df['Glucose']<=70, 'NewGlucose']='Low'
df.loc[(df['Glucose']>70) & (df['Glucose']<=99),'NewGlucose']='Normal'
df.loc[(df['Glucose']>99) & (df['Glucose']<=126),'NewGlucose']='Prediabetes'
df.loc[df['Glucose']>126, 'NewGlucose']='Diabetes'
```

##**Model Training after Feature engineering**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from scipy.interpolate import make_interp_spline

train_sizes = np.linspace(0.1, 0.9, 10)

models = {
    'KNN': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42)
}

train_accuracies = {name: [] for name in models}
test_accuracies = {name: [] for name in models}

for train_size in train_sizes:
    x_train_subset, _, y_train_subset, _ = train_test_split(x_train, y_train, train_size=train_size, random_state=42)

    for name, model in models.items():
        model.fit(x_train_subset, y_train_subset)

        train_pred = model.predict(x_train_subset)
        test_pred = model.predict(x_test)

        train_acc = accuracy_score(y_train_subset, train_pred)
        test_acc = accuracy_score(y_test, test_pred)

        train_accuracies[name].append(train_acc)
        test_accuracies[name].append(test_acc)

for name in models:
    final_train_acc = train_accuracies[name][-1]
    final_test_acc = test_accuracies[name][-1]
    print(f'Model: {name}, Final Training Accuracy: {final_train_acc:.4f}, Final Testing Accuracy: {final_test_acc:.4f}')

plt.figure(figsize=(12, 8))

colors = ['blue', 'green', 'red', 'purple']
markers_train = ['o', 's', '^', 'D']
markers_test = ['x', 'P', '*', 'v']

for i, (name, color) in enumerate(zip(models, colors)):
    x_smooth = np.linspace(train_sizes.min() * len(x_train), train_sizes.max() * len(x_train), 300)
    train_smooth = make_interp_spline(train_sizes * len(x_train), train_accuracies[name])(x_smooth)
    test_smooth = make_interp_spline(train_sizes * len(x_train), test_accuracies[name])(x_smooth)

    plt.plot(x_smooth, train_smooth, label=f'{name} - Training', color=color, marker=markers_train[i], markevery=30)
    plt.plot(x_smooth, test_smooth, label=f'{name} - Validation', linestyle='--', color=color, marker=markers_test[i], markevery=30)

plt.title('Effect of Training Set Size on Overfitting (KNN, Decision Tree, Random Forest, Logistic Regression)', fontsize=16)
plt.xlabel('Number of Training Samples', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=12)
plt.tight_layout()
plt.show()
```
